```python
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
```

```python
import numpy as np 
x = np.linspace(0,5,11) 
y = x**2 
```

```python
# Functional method 
plt.plot(x,y) 
plt.show()
```

```python
plt.plot(x,y,'r-') 
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title('title')
plt.show()
```

```python
plt.subplot(1,2,1) 
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.show()
```

```python
# Object Oriented Method 
```

```python
fig = plt.figure() 
axes = fig.add_axes([0.1,0.1,0.8,0.8]) 
axes.plot(x,y)
axes.set_xlabel('X Label') 
axes.set_ylabel('Y Label')
axes.set_title('Title')
plt.show() 
```

```python
fig = plt.figure() 
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) #Left Bottom, Width, Height 
axes1.plot(x,y)
axes1.set_title('Larger')
axes2.plot(y,x)
axes2.set_title('Smaller')
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2) 
#plt.tight_layout() # fix overlap
#axes.plot(x,y) 
for current_ax in axes: 
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2) 
axes[0].plot(x,y) 
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
# Figure Size and DPI 
fig = plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows= 2, ncols=1,figsize =(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('mypicture.png',dpi =200)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label = 'x squared') 
axes.plot(x,x**3,label = 'x cube')
axes.legend(loc = 0)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) 
ax.plot(x,y,color = 'purple',linewidth = 3, linestyle = '-',marker = 'o',markersize = 10,
       markerfacecolor = 'yellow',markeredgewidth = 3,markeredgecolor = 'green')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,linestyle = '--') 
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
